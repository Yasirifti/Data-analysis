```python
#Write a Python program to get the maximum number of confirmed, deaths, recovered and active cases 
#of Novel Coronavirus (COVID-19) Country wise.
small_data=df[["Country_Region", "Confirmed", 'Deaths', "Recovered", "Active"]]
small_data.groupby("Country_Region").max()
```

```python
# Write a Python program to get the total number of confirmed deaths and recovered people of Novel Coronavirus (COVID-19) cases
#Country/Region - Province/State wise
dr_data=df[['Country_Region', 'Province_State', 'Deaths', 'Recovered']]
dr_data_C =  dr_data.groupby('Country_Region').sum()
dr_data_R=  dr_data.groupby('Province_State').sum()
```

```python
#Write a Python program to create a plot (lines) of total deaths, confirmed, recovered and active cases Country wise
#where deaths greater than 150.
df2=df[['Country_Region', 'Deaths', 'Confirmed', 'Recovered', 'Active']]
death_data = df2[df2['Deaths']>150]
death_data_sum =  death_data.groupby('Country_Region').sum()
death_data_sum
```
